```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
covid= pd.read_csv('COVID-19.csv')
```

```python
covid.info()
```

```python
covid.head(23)
```

```python
plt.style.use('seaborn-darkgrid')
fig = plt.figure()
ax = plt.subplot()
ax.barh(covid['Name of State / UT'],covid['Death'],label='Deaths')
plt.title('COVID_19')
plt.xlabel('Deaths')
plt.ylabel('States')
plt.legend()
plt.rcParams["figure.figsize"] = (10,9)
fig.savefig('covid-deaths')
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
fig = plt.figure()
ax = plt.subplot()
ax.barh(covid['Name of State / UT'],covid['Total Confirmed cases (Indian National)'],label='Confirmed-cases')
ax.barh(covid['Name of State / UT'],covid['cured/dis'],label='cured/recoverd')
ax.barh(covid['Name of State / UT'],covid['Death'],label='Deaths')
plt.rcParams["figure.figsize"] = (8,8)
plt.title('COVID_19 data from https://www.mohfw.gov.in/')
plt.xlabel('Confirmed cases ')
plt.ylabel('States')
plt.legend()
#fig.savefig('covid-confirmed-cases')
plt.show()
```
